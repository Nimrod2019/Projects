```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Turning CSV to dataframe
Instacart=pd.read_csv('InstacartOrdersByDepartment.csv')
#Printing out the first few rows of the dataset.
Instacart.head()
```

```python
"""The follwong code will generate graphs of all departments in the grocery store found in the dataset.
There are 21 graphs because there are a totalof 21 departments. 
For each graph, we plot the hours of the day aginst the number of orders each hour per department. This waywe can see at what 
hour the most items in the depatment were ordered. 
"""
department_names=Instacart.department.unique()
for department_x in department_names:
    department_df=Instacart[Instacart.department==department_x]
    department_df.plot(x='order_hour_of_day',y='num_orders_hour', kind='bar')
    plt.ylabel('Number of Orders')
    plt.xlabel('Hour of Day')
    plt.title(department_x +' orders per hour')
    plt.show()
```

```python
"""It is difficult to compare sales per hour over departments so we can go one step futher and provide some statistics 
for the orders from each department. This way by looking at some numbers, maybe the graphs will make more sense.
"""
Instacart.groupby('department').describe()
```
